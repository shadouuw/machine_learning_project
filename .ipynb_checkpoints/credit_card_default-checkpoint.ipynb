** Analyse par rapport au nombre de comptes deault le moins prochain :

```python
g=sns.catplot(x ="Y", hue='X2',
kind ="count",data = dataset,legend_out=False)
plt.title('Number of default accounts in the next month \n by Gender', fontsize=15)
locs, labels=plt.xticks()
new_xticks = ['Not defaulted', 'Defaulted']
plt.xticks(locs, new_xticks,fontsize=15)
plt.xlabel('')
plt.ylabel('Count', fontsize=15)
new_labels = ['Male', 'Female']
for t, l in zip(g._legend.texts, new_labels): t.set_text(l)



```

```python
#pour cette column on voit que seulment 4 Types ont a sens les autres sont
#useless
g=sns.catplot(x ="Y", hue='X3',
kind ="count",data = dataset,legend_out=False)
plt.title('Number of default accounts in the next month \n by Education', fontsize=15)
locs, labels=plt.xticks()
new_xticks = ['Not defaulted', 'Defaulted']
plt.xticks(locs, new_xticks,fontsize=15)
plt.xlabel('')
new_labels = ['Graduate school','University','High School', 'Others']
for t, l in zip(g._legend.texts, new_labels): t.set_text(l)
```

Now lets draw the histograms of the different features :

```python
# set up the figure size
plt.style.use('seaborn')
plt.rcParams['figure.figsize'] = (20, 10)

# make subplots (2 rows and 2 columns)
fig, axes = plt.subplots(nrows = 2, ncols = 3)

# Specify the features of interest
# The pay status features are columns 6 through 11 inclusive
pay_features = dataset.columns[5:11]

# Use variables as x-axis labels and 'Counts' as y-axis labels
xaxes = pay_features


# draw histograms
# we can then use enumerate on axes.
axes = axes.ravel()
for idx, ax in enumerate(axes):
    ax.hist(dataset[pay_features[idx]].dropna(), bins=11)
    ax.set_ylabel('Counts', fontsize=20)
    ax.tick_params(axis='both', labelsize=15)
    ax.set_title("Histogram of {} variable".format(xaxes[idx]),fontsize=15)

plt.tight_layout()
plt.show()
```

en eleminant X6 tout les X7,8,9,10,11 on des valeurs manquant pour X=1

```python
# set up the figure size
plt.style.use('seaborn')
plt.rcParams['figure.figsize'] = (20, 10)

# make subplots (2 rows and 2 columns)
fig, axes = plt.subplots(nrows = 2, ncols = 3)

# Specify the features of interest
# The pay status features are columns 6 through 11 inclusive
bill_features = dataset.columns[11:17]

# Use variables as x-axis labels and 'Counts' as y-axis labels
xaxes = bill_features


# draw histograms
# we can then use enumerate on axes.
axes = axes.ravel()
for idx, ax in enumerate(axes):
    ax.hist(dataset[bill_features[idx]].dropna(), bins=11 , color="red")
    ax.set_ylabel('Counts', fontsize=20)
    ax.tick_params(axis='both', labelsize=15)
    ax.set_title("Histogram of {} variable".format(xaxes[idx]),fontsize=15)

plt.tight_layout()
plt.show()
```
